```python
import pandas as pd
```

```python
population_data= pd.read_csv('population_data.csv')
```

```python
population_data
```

```python
population_data = population_data.drop(['country', 'continent', 'iso_alpha'], axis=1)
population_data
```

```python
Y = population_data.gdpPercap # Dependent variable (target)
Y = Y.values
Y
```

```python
X = population_data.drop(['gdpPercap'], axis=1) # independent variables
X = X.values
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
pop_model = linear_model.LinearRegression()
```

```python
pop_model.fit(X_train, Y_train)
```

```python
Y_pred = pop_model.predict(X_test)
```

```python
print('Coefficients:', pop_model.coef_)
print('Intercept:', pop_model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_test, Y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_test, Y_pred))
```
